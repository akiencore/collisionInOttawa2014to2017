```python
print("make your plot outputs appear and be stored within the notebook")
%matplotlib  inline               
import pandas as pd
import numpy as np
from pykml import parser

print("do not use html representation in pandas object")
pd.set_option('html', False)    
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)
```

```python
xls_2014 = pd.read_excel('2014collisionsfinal.xls', sheetname='2014_all_records')
xls_2014.drop(['X', 'Y'], axis=1, inplace=True)

with open('2014collisionsfinal.kml') as f:
  folder = parser.parse(f).getroot().Document.Folder

longitude = list()
latitude = list()

for pm in folder.Placemark:
    coordinate = str(pm.Point.coordinates).split(",")
    longitude.append(float(coordinate[0])) 
    latitude.append(float(coordinate[1]))

xls_2014['longitude'] = longitude
xls_2014['latitude'] = latitude
xls_2014['Record'] = xls_2014['Record'] + 2014000000

xls_2014.to_csv('collision2014.csv', index=False)

xls_2014
```

```python
from urllib import urlencode, urlopen
import simplejson as json

xls_2016 = pd.read_excel('2016collisionsfinal.xls', sheetname='2016_all_records')
xls_2016.drop(['X', 'Y'], axis=1, inplace=True)

with open('2016collisionsfinal.kml') as f:
  folder = parser.parse(f).getroot().Document.Folder

longitude = list()
latitude = list()

for pm in folder.Placemark:
    coordinate = str(pm.Point.coordinates).split(",")
    longitude.append(float(coordinate[0])) 
    latitude.append(float(coordinate[1]))
    
print(len(xls_2016))
print(len(longitude))

location = list()
n = len(longitude)

for pm in folder.Placemark:
    location.append(pm.name)

for i in range(n):
    if (xls_2016)['Location'][i] != location[i]:
        print(i)
        print("xls_2016-%d: %s " % (i-1, (xls_2016)['Location'][i-1]))
        print("kml_2016-%d: %s\n " % (i-1, location[i-1]))
        print("xls_2016-%d: %s " % (i,   (xls_2016)['Location'][i]))
        print("kml_2016-%d: %s\n " % (i,   location[i]))
        print("xls_2016-%d: %s " % (i+1, (xls_2016)['Location'][i+1]))
        print("kml_2016-%d: %s\n " % (i+1, location[i+1]))
        break

lostLocation = 'VIEILLE-CASERNE PRIV, Ottawa, ON, CA'
lostLocation = urlencode({'address' : lostLocation})
url = "http://maps.google.com/maps/api/geocode/json?"+lostLocation+"&sensor=false"

tmpLocation = urlopen(url)
tmpjson = json.loads(tmpLocation.read())
tmplon = tmpjson.get(u'results')[0][u'geometry'][u'location'][u'lng']    # u -> unicode string
tmplat = tmpjson.get(u'results')[0][u'geometry'][u'location'][u'lat']

longitude.insert(i, tmplon)
latitude.insert(i, tmplat)

xls_2016['longitude'] = longitude
xls_2016['latitude'] = latitude
xls_2016['Record'] = xls_2016['Record'] + 2016000000

print xls_2016.ix[i]

xls_2016.to_csv('collision2016.csv', index=False)

xls_2016
```
